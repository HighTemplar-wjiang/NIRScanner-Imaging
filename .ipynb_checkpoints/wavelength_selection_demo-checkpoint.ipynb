```python
import os
import re
import pickle
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.patheffects as pe
from mpl_toolkits.axes_grid1 import make_axes_locatable
import seaborn as sns

import sys
sys.path.append(".") # Adds current directory to python modules path.
from NIRSignal.NIRSignal import *
```

```python
# Params.
mode_export = False

# Default font size.
font = {"size": 16}
mpl.rc('font', **font)

# Colormap.
# img_cmap = "binary"
img_cmap = "viridis"
# img_cmap = "magma"

ax_label_params = {
    "fontsize": 20,
    "fontstyle": "italic",
    "fontweight": "bold"
}
tick_label_params = {
    "fontsize": 18,
    "fontweight": "bold"
}

ax_label_params_small = {
    "fontsize": 14,
    "fontstyle": "italic",
    "fontweight": "bold"
}
tick_label_params_small = {
    "fontsize": 12,
    "fontweight": "bold"
}

ax_label_params_large = {
    "fontsize": 30,
    "fontstyle": "italic",
    "fontweight": "bold"
}
tick_label_params_large = {
    "fontsize": 28,
    "fontweight": "bold"
}

# Percentiles.
low_percentile = 2
high_percentile = 98
```

```python
# Load data.
data_path = "./data/"
pixel_size = 1
re_file = re.compile("demo_EUREKA_papermarker_15_layers_{:d}mm_01".format(pixel_size))

all_filenames = os.listdir(data_path)

all_data_raw_3d = []
for filename in all_filenames:
    
    re_result = re_file.match(filename)
    
    if re_result:
        # Filename matched. 
        print(filename)

        with open(os.path.join(data_path, filename), "rb") as f:
            data_raw_2d = pickle.load(f)

        all_data_raw_3d.append(data_raw_2d)
        
        break

# Get wavelengths.
wavelength_list = all_data_raw_3d[0][0][0]["wavelength"]
wavelength_list = np.array(wavelength_list)

# Print raw data shape. 
all_data_raw_3d = np.array(all_data_raw_3d)
print(np.shape(all_data_raw_3d))
```

```python
# Constructing data.

# Axis 0: scan 
# Axis 1: y
# Axis 2: x
# Axis 3: wavelength

x_min = 0
x_max = all_data_raw_3d.shape[2]
y_min = 0
y_max = all_data_raw_3d.shape[1]

all_intensity_raw_3d = []
all_data_processed_3d = [] 
for data_raw_2d in all_data_raw_3d:
    # Processing each scan.
    
    intensity_raw_2d = []
    data_processed_2d = []
    for idx_y in range(y_min, y_max):

        # Init.
        intensity_raw_1d = []
        data_processed_1d = []
        data_raw_1d = data_raw_2d[idx_y]
        for idx_x in range(x_min, x_max):

            # Get single scan data.
            data_raw = data_raw_1d[idx_x]

            # Get raw spectrum. 
            intensity_raw = np.array(data_raw["intensity"])
            reference_raw = np.array(data_raw["reference"])
            
            # Processing.
            wavelength_list_processed, processed_spectrum = process_signal(wavelength_list, intensity_raw, reference_raw, reference_raw, 
                                                                           selected_indexes=None, decimate_factor=1, absorbance_mode=False) 
            
            # End of pixel. 
            intensity_raw_1d.append(intensity_raw)
            data_processed_1d.append(processed_spectrum)
        
        # End of line, append.
        intensity_raw_2d.append(intensity_raw_1d)
        data_processed_2d.append(data_processed_1d)
    
    # Convert to numpy array and append.
    intensity_raw_2d = np.array(intensity_raw_2d)
    data_processed_2d = np.array(data_processed_2d)
    all_intensity_raw_3d.append(intensity_raw_2d)
    all_data_processed_3d.append(data_processed_2d)
        
all_intensity_raw_3d = np.array(all_intensity_raw_3d)
all_data_processed_3d = np.array(all_data_processed_3d)
all_data_labels_2d = np.array([(x, y) for y in range(data_processed_2d.shape[0]) 
                            for x in range(data_processed_2d.shape[1])])

print(all_intensity_raw_3d.shape)
print(all_data_processed_3d.shape)
print(all_data_labels_2d[:3])
```

```python
# Calculate single wavelength case - raw.
target_wavelength = 1302
idx_target_wavelength = np.argmin(np.abs(wavelength_list - target_wavelength))
print("target wavelength: {:.2f}".format(wavelength_list[idx_target_wavelength]))

# Slice data.
data_slice_single_wavelength = all_intensity_raw_3d[0, :, :, idx_target_wavelength]
img_single_wavelength = normalize_array(data_slice_single_wavelength, low_percentile=low_percentile, high_percentile=high_percentile)

# Show image.
fig, ax = plt.subplots(dpi=300)
ax.set_xlabel("x (mm)", **ax_label_params)
ax.set_ylabel("y (mm)", **ax_label_params)
ax.set_xticks([0, img_single_wavelength.shape[1]])
ax.set_yticks([0, img_single_wavelength.shape[0]])
ax.set_xticklabels([0, img_single_wavelength.shape[1] * pixel_size], **tick_label_params)
ax.set_yticklabels([0, img_single_wavelength.shape[0] * pixel_size], **tick_label_params)
ax_img = ax.imshow(img_single_wavelength, extent=(0, img_single_wavelength.shape[1], 0, img_single_wavelength.shape[0]), cmap=img_cmap)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
fig.colorbar(ax_img, cax=cax, ticks=[0, 0.5, 1.0])

# Plot histogram.
# Get histogram results.
# Without ground truth -- left wing vs. right wing.
num_bins = 20
total_counts = np.prod(img_single_wavelength.shape)
fig, ax = plt.subplots(figsize=(7, 5), dpi=300)
ax.set_ylim([0, 0.2])
ax.set_yticks([0, 0.05, 0.10, 0.15, 0.20])
ax.set_xlabel("Image intensity values", **ax_label_params)
ax.set_ylabel("Normalized histogram", **ax_label_params)
hist_counts_left, bins_left = np.histogram(img_single_wavelength[img_single_wavelength < 0.5], bins=num_bins, range=(0, 1))
hist_counts_right, bins_right = np.histogram(img_single_wavelength[img_single_wavelength >= 0.5], bins=num_bins, range=(0, 1))
print("DM = {:.3f}".format(np.mean(img_single_wavelength[img_single_wavelength >= 0.5]) - 
                           np.mean(img_single_wavelength[img_single_wavelength < 0.5])))
ax.hist(bins_left[:-1], bins=bins_left, weights=hist_counts_left / total_counts, 
        range=(0, 1), color="#64c9f5", ec="black", label="Values < 0.5")
ax.hist(bins_right[:-1], bins=bins_right, weights=hist_counts_right / total_counts, 
        range=(0, 1), color="#f57e76", ec="black", label="Values > 0.5")
ax.legend(loc="upper left")
```

```python
# Step-wise additive algorithm.
from progressbar import ProgressBar
from sklearn.metrics import mean_squared_error

print((low_percentile, high_percentile))

bar = ProgressBar(max_value=len(wavelength_list))
bar.update(0)

def calculate_divergence(normalized_array):
    mean_high_end = np.mean(normalized_array[normalized_array > 0.5])
    mean_low_end = np.mean(normalized_array[normalized_array < 0.5])
    
    return mean_high_end - mean_low_end

# Slice current layer data.
intensity_raw_current_layer = all_data_processed_3d[0, :, :, :]
print("Using smoothed data.")

# Adding wavelength one by one.
idx_current_wavelengths_div_raw = []
idx_current_wavelength_pool_div_raw = list(range(len(wavelength_list)))
wavelength_subset_list_div_raw = []
mse_list_wavelength_subsets_div_raw = []
div_list_wavelength_subsets_div_raw = []
image_list_current_layer_wavelengths_div_raw = []
for idx_epoch in range(len(wavelength_list)):

    # Calculate different means by removing one wavelength, each time, respectively.
    div_list_current_epoch_div = []

    # Init current subset. 
    for idx_add_wavelength in range(len(idx_current_wavelength_pool_div_raw)):

        # Add one wavelength.
        idx_wavelengths_added_div = idx_current_wavelengths_div_raw + [idx_current_wavelength_pool_div_raw[idx_add_wavelength]]

        # Get the image - div algo.
        intensity_slice_current_subset_div_raw = intensity_raw_current_layer[:, :, idx_wavelengths_added_div]
        intensity_slice_current_subset_div_mean = np.mean(intensity_slice_current_subset_div_raw, axis=-1)
        intensity_slice_current_subset_div_norm_raw = normalize_array(intensity_slice_current_subset_div_mean,
                                                                      low_percentile=low_percentile, high_percentile=high_percentile)

        # Calcualte divs.
        div_current_subset_div = calculate_divergence(intensity_slice_current_subset_div_norm_raw)
        div_list_current_epoch_div.append(div_current_subset_div)

    # Find the index with max div. 
    idx_add_wavelength_div = np.argmax(div_list_current_epoch_div)

    # Store the div.
    div_list_wavelength_subsets_div_raw.append(div_list_current_epoch_div[idx_add_wavelength_div])

    # Adding that wavelength.
    idx_current_wavelengths_div_raw += [idx_current_wavelength_pool_div_raw[idx_add_wavelength_div]]

    # Removing that wavelength from the pools.
    idx_current_wavelength_pool_div_raw = np.delete(idx_current_wavelength_pool_div_raw, idx_add_wavelength_div)

    # Get the images.
    intensity_slice_current_subset_div_norm_raw = normalize_array(
        np.mean(intensity_raw_current_layer[:, :, idx_current_wavelengths_div_raw], axis=-1),
        low_percentile=low_percentile, high_percentile=high_percentile)
    image_list_current_layer_wavelengths_div_raw.append(intensity_slice_current_subset_div_norm_raw)
    
    bar.update(idx_epoch+1)

print("Done")
```

```python
# Visualize selected wavelengths.
# Plot progress.
idx_max_div = np.argmax(div_list_wavelength_subsets_div_raw)
x_array = list(range(1, len(wavelength_list)+1))

#### DIV.
fig, ax = plt.subplots(figsize=(9, 7), dpi=300)
plt.setp(ax.get_xticklabels() + ax.get_yticklabels(), **tick_label_params_large)
ax.set_xticks([0, 100, 200])
# ax.set_ylim([0.4, 0.8])
ax.set_yticks([0.40, 0.60, 0.80])
# ax.set_yticklabels([0.40, 0.60, 0.80], **tick_label_params_large)
ax.set_xlabel("Number of iterations", **ax_label_params_large)
ax.set_ylabel("Mean difference", **ax_label_params_large)
line_div = ax.plot(div_list_wavelength_subsets_div_raw, lw=2, label="Mean difference", color="#f57e76")
ax.scatter(idx_max_div, div_list_wavelength_subsets_div_raw[idx_max_div], 
           marker="o", s=30, linewidth=0.5, edgecolor="black", color="#f57e76", zorder=100)
fig.tight_layout()
```
